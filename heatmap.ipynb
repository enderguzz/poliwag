```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import seaborn as sns
%matplotlib inline
```

```python
# load the dataset
df = pd.read_csv('Spotify-2000.csv')
```

```python
print(df.info())
```

```python
# average valence(poritiveness) of top 2000 songs
Average_of_Valence = df['Valence'].mean()
print(Average_of_Valence)
```

```python
# average valence(positiveness) per year
average = df.groupby(by='Year')['Valence'].mean()
print(average)
```

```python
# the average valence over the years with line plot
plt.figure(figsize=(10, 5))
plt.plot(average)
plt.ylabel('Average of Valance')
plt.xlabel('Year')
plt.grid(linestyle=':')
plt.show()
```

```python
# Becasuse of the intersection square of Speechiness and Popularity is above 0, it can be said that there is a slight linear relationship.
# Because of the intersection square of BMP and Popularity is very close to 0, it can be said that there is no relationship between them.
plt.figure(figsize = (16,5))
sns.heatmap(df.corr(), cmap="coolwarm", annot=True, )
plt.show()
```
